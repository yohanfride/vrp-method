```python
import numpy as np
import pandas as pd
import json

n = 50
customer = [i for i in range(1, n + 1)]
node = [0] + customer
arcos = [(i,j) for i in node for j in node if i != j]
f = open('pelanggan.json')
data = json.load(f)
# get customer demand and location data
df = pd.DataFrame(data)
# print(df)
# get distance data
distance = np.loadtxt('distance.txt')
vehicle_capacity = 250
```

```python
def saving(data_distance,demand_data,node,depot,i,capacity):
    S = [[0 for i in range(n)]for i in range(n)]
    max_saving = 0
    res = [0,0]
    for j in range(1,n+1):
        if (j in node) and (capacity - demand_data[j] > 0) and (j != i):
            S1 = data_distance[i][0] + data_distance[0][j] - data_distance[i][j]
#             print("i="+str(i)+", j="+str(j)+" --> "+str(data_distance[i][0])+" + "+str(data_distance[0][j])+" - "+str(data_distance[i][j])+" = "+str(S1))
            S2 = data_distance[0][i] + data_distance[j][0] - data_distance[j][i]
#             print("i="+str(j)+", j="+str(i)+" --> "+str(data_distance[j][0])+" + "+str(data_distance[0][i])+" - "+str(data_distance[j][i])+" = "+str(S2))
            if(max_saving<S1):
                max_saving = S1
                res = [j,1]
            if(max_saving<S2):
                max_saving = S2
                res = [j,-1]        
#     print(max_saving)
#     print(res)
    return res
```

```python
def farnode(data_distance,node,depot):
    max_range = 0
    res = 0
    for j in range(1,n+1):
        if (j in node):
            if(max_range<data_distance[depot][j]):
                max_range = data_distance[depot][j]
                res = j
#     print(res)
    return res    
```

```python
nodes = node[:]
route = 1
routes = []
croute = []

###Menghitung Waktu
import time
start_timer = time.time()

#Titik Terjauh
i = farnode(distance, node, 0)
capacity = vehicle_capacity
croute.append(i)
nodes.remove(i)
while True:
    if len(nodes) == 1:
        croute.append(0)
        croute.insert(0,0)
        routes.append(croute)
        break
    S = saving(distance,df.demand,nodes,0,i,capacity)
    j = S[0]
    t = S[1]
    if(t==1):
        croute.append(j)
    elif(t==-1):
        croute.insert(0,j)
    print("Rute : "+str(route))
    print(croute)
    print(nodes)
    print("---------------------------------------------------")
    if(j != 0):
        nodes.remove(j)
        capacity-=df.demand[j]
        i = j
    else:
        croute.append(0)
        croute.insert(0,0)
        routes.append(croute)
        route+=1
        croute = []
        i = farnode(distance, nodes, 0)
        capacity = vehicle_capacity
        croute.append(i)
        nodes.remove(i)
        if len(nodes) == 1:
            croute.append(0)
            croute.insert(0,0)
            routes.append(croute)
            break

###Menghitung Waktu
done_timer = time.time()
elapsed = done_timer - start_timer
```
