```python
import numpy as np
import pandas as pd
import json

n = 15
customer = [i for i in range(1, n + 1)]
node = [0] + customer
arcos = [(i,j) for i in node for j in node if i != j]
f = open('pelanggan.json')
data = json.load(f)
# get customer demand and location data
df = pd.DataFrame(data)
# print(df)
# get distance data
distance = np.loadtxt('distance.txt')
vehicle_capacity = 250
```

```python
def saving(data_distance,node,depot,i):
    S = [[0 for i in range(n)]for i in range(n)]
    max_saving = 0
    res = [0,0]
    for j in range(1,n+1):
        if (j in node) and (j != i):
            S1 = data_distance[i][0] + data_distance[0][j] - data_distance[i][j]
            S2 = data_distance[0][i] + data_distance[j][0] - data_distance[j][i]
            if(max_saving<S1):
                max_saving = S1
                res = [j,1]
            if(max_saving<S2):
                max_saving = S2
                res = [j,-1]        
    return res
```

```python
def farnode(data_distance,node,depot):
    max_range = 0
    res = 0
    for j in range(1,n+1):
        if (j in node):
            if(max_range<data_distance[depot][j]):
                max_range = data_distance[depot][j]
                res = j
    return res    

def firstsaving(data_distance,node,depot):
    max_saving = 0
    res = [0,0]
    for j in range(1,n+1):
        for i in range(1,n+1):
            if (j in node) and (i in node) and (j != i):
                S1 = data_distance[i][0] + data_distance[0][j] - data_distance[i][j]
                S2 = data_distance[0][i] + data_distance[j][0] - data_distance[j][i]
                if(max_saving<S1):
                    max_saving = S1
                    res = [i,j]
                if(max_saving<S2):
                    max_saving = S2
                    res = [j,i]
    return res
```

```python
nodes = node[:]
route = 1
routes = []
croute = []

###Menghitung Waktu
import time
start_timer = time.time()

#Titik Terjauh
# i = farnode(distance, node, 0)
S = firstsaving(distance, node, 0)
capacity = vehicle_capacity
croute.append(S[0])
nodes.remove(S[0])
capacity-=df.demand[S[0]]
croute.append(S[1])
nodes.remove(S[1])
capacity-=df.demand[S[1]]
i = S[0]
while True:
    if len(nodes) == 1:
        croute.append(0)
        croute.insert(0,0)
        routes.append(croute)
        break
    S = saving(distance,nodes,0,i)
    j = S[0]
    t = S[1]
    if(capacity - df.demand[j] > 0):
        if(t==1):
            croute.append(j)
        elif(t==-1):
            croute.insert(0,j)
        print("Rute : "+str(route))
        print(croute)
        print(nodes)
        print("---------------------------------------------------")
    else:
        j = 0
        
    if(j != 0):
        nodes.remove(j)
        capacity-=df.demand[j]
        i = j
    else:
        croute.append(0)
        croute.insert(0,0)
        routes.append(croute)
        route+=1
        croute = []
#         i = S[0]
        i = farnode(distance, nodes, 0)
        capacity = vehicle_capacity
        capacity-=df.demand[i]
        croute.append(i)
        nodes.remove(i)
        if len(nodes) == 1:
            croute.append(0)
            croute.insert(0,0)
            routes.append(croute)
            break

###Menghitung Waktu
done_timer = time.time()
elapsed = done_timer - start_timer
```

```python
totals = 0
for r in range(len(routes)):
    print("Rute "+str(r+1))
    total = 0
    permintaan = 0
    for l in range(1,len(routes[r])) :
        total+=distance[routes[r][l-1]][routes[r][l]]
        permintaan+=df.demand[routes[r][l]]
    print(str(routes[r])+" -> "+str(total)+" -> "+str(permintaan))
    totals+=total 
print(totals)
```
